# À propos de l'ensemble de données
## Marketing bancaire
### Résumé :
Les données sont liées aux campagnes de marketing direct (appels téléphoniques) d'une institution bancaire portugaise. L'objectif de la classification est de prédire si le client souscrira un dépôt à terme (variable y).

### Informations sur l'ensemble de données :
les données sont liées aux campagnes de marketing direct d'une institution bancaire portugaise. Les campagnes marketing étaient basées sur des appels téléphoniques. Souvent, plusieurs contacts avec le même client étaient nécessaires pour savoir si le produit (dépôt bancaire à terme) serait (« oui ») ou non (« non ») souscrit.

## Informations sur les attributs :
### Données clients bancaires :
Âge (numérique)
Emploi : type d'emploi (catégorique : 'admin.', 'ouvrier', 'entrepreneur', 'femme de ménage', 'gestion', 'retraité', 'indépendant', 'services', 'étudiant', ' technicien', 'chômeur', 'inconnu')
Matrimonial : état civil (catégorique : 'divorcé', 'marié', 'célibataire', 'inconnu' ; note : 'divorcé' signifie divorcé ou veuf)
Éducation (catégorique : « de base.4 ans », « de base.6 ans », « de base.9 ans », « lycée », « illettré », « cours.professionnel », « diplôme.universitaire », « inconnu »)
Défaut : le crédit est-il en défaut ? (catégorique : "non", "oui", "inconnu")
Logement : existe-t-il un prêt logement ? (catégorique : "non", "oui", "inconnu")
Prêt : avez-vous un prêt personnel ? (catégorique : "non", "oui", "inconnu")

### En lien avec le dernier contact de la campagne en cours :
Contact : type de communication du contact (catégorique :
« cellulaire », « téléphone »)
Mois : mois du dernier contact de l'année (catégorique : 'jan', 'feb', 'mar',
…, 'nov', 'dec')
Day_of_week : dernier jour de contact de la semaine (catégorique :
'lundi', 'mardi', 'mercredi', 'jeu', 'ven')
Durée : durée du dernier contact, en secondes (numérique).

### Autres attributs :
Campagne : nombre de contacts effectués lors de cette campagne et pour
ce client (numérique, inclut le dernier contact)
Pdays : nombre de jours écoulés après que le client a été
contacté pour la dernière fois lors d'une campagne précédente (numérique ; 999 signifie que le client n'a pas été
contacté auparavant)
Précédent : nombre de contacts effectués avant cette campagne et pour
ce client (numérique)
Poutcome : résultat de la campagne marketing précédente (catégorique :
"échec", "inexistant", "succès")

### Attributs du contexte social et économique
Emp.var.rate : taux de variation de l'emploi - indicateur trimestriel
(numérique)
Cons.price.idx : indice des prix à la consommation - indicateur mensuel (numérique)
Cons.conf.idx : indice de confiance des consommateurs - indicateur mensuel
(numérique)
Euribor3m : taux euribor 3 mois - indicateur journalier (numérique)
Nr.employed: nombre d'employés - indicateur trimestriel (numérique)

### Variable de sortie (cible souhaitée) :
y - le client a-t-il souscrit un dépôt à terme ? (binaire : 'oui', 'non')

## Étapes d'analyse :
Analyse des informations sur les attributs.
Apprentissage automatique (régression logistique, KNN, SVM, arbre de décision,
forêt aléatoire, Naive Bayes)
Apprentissage profond (ANN)

```python
var_categorielle = ['metier', 'situation_matrimoniale', 'education', 'defaut_credit', 'logement', 'pret', 'contact', 'mois', 'jour_de_semaine', 'resultat_campagne_precendente']

for col in var_categorielle :
    print(col, ":", data[col].unique())
    print('\n')
```

```python
# Traductions des valeurs
education_nouvelles_valeurs = ["inconnue", "illettré", "basic.4ans", "basic.6ans", "basic.9ans", "lycée", "cours_professionnel", "diplôme_universitaire"]
education_valeurs = ["unknown", "illiterate", "basic.4y", "basic.6y", "basic.9y", "high.school", "professional.course", "university.degree"]

metier_nouvelles_valeurs = ["inconnue", "admin.", "ouvrier", "entrepreneur", "femme_de_ménage", "gestion", "retraité", "indépendant", "services", "étudiant", "technicien", "chômeur"]
metier_valeurs = ["unknown", "admin.", "blue-collar", "entrepreneur", "housemaid", "management", "retired", "self-employed", "services", "student", "technician", "unemployed"]

situation_matrimoniale_nouvelles_valeurs = ["inconnue", "divorcé", "marié", "célibataire"]
situation_matrimoniale_valeurs = ["unknown", "divorced", "married", "single"]

contact_nouvelles_valeurs = ["cellulaire", "téléphone"]
contact_valeurs = ["cellular", "telephone"]

mois_nouvelles_valeurs = ['avr', 'août', 'déc', 'juil', 'juin', 'mar', 'mai', 'nov', 'oct', 'sep']
mois_valeurs = ['apr', 'aug', 'dec', 'jul', 'jun', 'mar', 'may', 'nov', 'oct', 'sep']

jour_semaine_nouvelles_valeurs = ['ven', 'lun', 'jeu', 'mar', 'mer']
jour_semaine_valeurs = ['fri', 'mon', 'thu', 'tue', 'wed']

defaut_credit_nouvelles_valeurs = ["oui", "non", "inconnue"]
defaut_credit_valeurs = ["yes", "no", "unknown"]

logement_nouvelles_valeurs = ['non', 'oui', 'inconnue']
logement_valeurs = ['no', 'yes', 'unknown']

pret_nouvelles_valeurs = ["non", "oui", "inconnue"]
pret_valeurs = ['no', 'yes', 'unknown']

resultat_campagne_precedente_valeurs = ['nonexistent', 'failure', 'success']
resultat_campagne_precedente_nouvelles_valeurs = ['existe_pas', 'echec', 'reussite']

y_nouvelles_valeurs = ["oui", "non"]
y_precedente_valeur = ["yes", "no"] 
```

```python
# Remplacement des valeurs dans le DataFrame
data['jour_de_semaine'].replace(jour_semaine_valeurs, jour_semaine_nouvelles_valeurs, inplace=True)
data['mois'].replace(mois_valeurs, mois_nouvelles_valeurs, inplace=True)
data['contact'].replace(contact_valeurs, contact_nouvelles_valeurs, inplace=True)
data['metier'].replace(metier_valeurs, metier_nouvelles_valeurs, inplace=True)
data['situation_matrimoniale'].replace(situation_matrimoniale_valeurs, situation_matrimoniale_nouvelles_valeurs, inplace=True)
data['education'].replace(education_valeurs, education_nouvelles_valeurs, inplace=True)
data['defaut_credit'].replace(defaut_credit_valeurs, defaut_credit_nouvelles_valeurs, inplace=True)
data['logement'].replace(logement_valeurs, logement_nouvelles_valeurs, inplace=True)
data['pret'].replace(pret_valeurs, pret_nouvelles_valeurs, inplace=True)
data['resultat_campagne_precendente'].replace(resultat_campagne_precedente_valeurs, resultat_campagne_precedente_nouvelles_valeurs, inplace=True)
data["y"].replace(y_precedente_valeur, y_nouvelles_valeurs, inplace=True)
```

L'analyse univariée des variables catégorielles que nous avons effectuée est une approche pour comprendre la distribution des différentes catégories par rapport à la variable cible "y" (souscription à un dépôt à terme).

### 1. **metier (type d'emploi) :**
   - La plupart des clients semblent être dans les catégories "admin." et "technicien".
   - Les étudiants ont une tendance plus forte à souscrire à un dépôt à terme par rapport à d'autres catégories.

### 2. **situation_matrimoniale (état civil) :**
   - La majorité des clients sont mariés, suivis par des célibataires.
   - Les célibataires ont une légère tendance à souscrire davantage.

### 3. **education (niveau d'éducation) :**
   - La plupart des clients ont un niveau d'éducation de base ou universitaire.
   - Les clients avec un diplôme universitaire semblent avoir une tendance plus forte à souscrire.

### 4. **defaut_credit (crédit en défaut) :**
   - La plupart des clients n'ont pas de défaut de crédit.
   - Les clients sans défaut de crédit ont une tendance plus forte à souscrire.

### 5. **logement (prêt logement) :**
   - La plupart des clients ne disposent pas d'un prêt logement.
   - Les clients sans prêt logement ont une tendance plus forte à souscrire.

### 6. **pret (prêt personnel) :**
   - La plupart des clients n'ont pas de prêt personnel.
   - Les clients sans prêt personnel ont une tendance plus forte à souscrire.

### 7. **contact (type de communication) :**
   - La communication cellulaire semble plus fréquente que la communication téléphonique.
   - Les clients contactés par téléphone semblent avoir une légère tendance à souscrire davantage.

### 8. **mois (mois du dernier contact) :**
   - Les contacts semblent répartis sur plusieurs mois.
   - Il n'y a pas de tendance claire en fonction du mois.

### 9. **jour_de_semaine (jour de la semaine du dernier contact) :**
   - Les contacts sont également répartis sur tous les jours de la semaine.
   - Il n'y a pas de tendance claire en fonction du jour de la semaine.

### 10. **resultat_campagne_precendente (résultat de la campagne marketing précédente) :**
   - La plupart des clients n'ont pas de résultat de campagne précédente.
   - Les clients avec un résultat de campagne précédente "succès" ont une tendance plus forte à souscrire.

### 11. **y (souscription à un dépôt à terme) :**
   - Il est essentiel de vérifier la distribution des classes dans la variable cible.
   - La classe "no" semble être plus fréquente, ce qui indique un déséquilibre dans les données.

Globalement, ces graphiques aident à identifier des tendances et des relations potentielles entre les variables catégorielles et la souscription à un dépôt à terme.

```python
# Analyse univariée des variables numériques
numerical_columns = ['age', 'campagne','duree_appel' , 'nombre_de_jour_ecoule', 'nombre_contact_precedent']

# Création d'un tableau de subplots
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(18, 14))

# Ajustement de l'espacement entre les sous-graphiques
plt.subplots_adjust(wspace=0.5, hspace=0.5)

# Itération sur les colonnes numériques
for i, col in enumerate(numerical_columns):
    # Sélection du subplot approprié
    ax = axes[i // 2, i % 2]
    
    # Tracé du graphique pour la souscription à un dépôt à terme
    sns.histplot(data[data['y'] == 'oui'][col], bins=30, kde=True, color='red', ax=ax)
    
    # Tracé du graphique pour la non souscription à un dépôt à terme
    sns.histplot(data[data['y'] == 'non'][col], bins=30, kde=True, color='#2E86C1', ax=ax)
    
    # Définition des étiquettes et du titre
    ax.set_xlabel(f'{col}')
    ax.set_ylabel('Count')
    ax.set_title(f'Distribution de {col}')

# Ajout de la légende à la figure
fig.legend(labels=['Souscription', 'Non souscription'], loc='upper right')

# Affichage du tableau de graphiques
plt.show()
```

Pour chaque variable numérique, nous comparons la distribution entre les clients qui ont souscrit à un dépôt à terme ("oui") et ceux qui n'ont pas souscrit ("non").

### 1. **Age :**
- **Observations :**
  - Les clients qui souscrivent à un dépôt à terme ont une répartition plus large en termes d'âge par rapport à ceux qui ne souscrivent pas.
  - On observe une concentration plus importante de clients plus jeunes parmi ceux qui souscrivent.

### 2. **duree_appel (Durée du dernier contact en secondes) :**
- **Observations :**
  - La durée du dernier contact semble significativement plus longue pour les clients qui souscrivent à un dépôt à terme.
- **Remarque importante :**
  - L'attribut durée affecte fortement la cible de sortie (par exemple, si
  durée=0 alors y='no'). Cependant, la durée n'est pas connue avant qu'un appel
  soit effectué. De plus, après la fin de l'appel, y est évidemment connu.
  Ainsi, cette donnée ne doit être incluse qu’à des fins de référence et
  doit être écartée si l’intention est d’avoir un
  modèle prédictif réaliste.


### 3. **campagne (Nombre de contacts effectués lors de cette campagne) :**
- **Observations :**
  - Les clients qui souscrivent à un dépôt à terme ont tendance à avoir été contactés un nombre plus faible de fois au cours de cette campagne.

### 4. **nombre_de_jour_ecoule (Nombre de jours écoulés après le dernier contact lors d'une campagne précédente) :**
- **Observations :**
  - La majorité des clients, qu'ils souscrivent ou non, n'ont pas été contactés lors d'une campagne précédente (pdays = 999).
  - Cependant, parmi ceux qui ont été contactés précédemment, il semble y avoir une distribution similaire entre les deux groupes.

### 5. **nombre_contact_precedent (Nombre de contacts effectués avant cette campagne) :**
- **Observations :**
  - Les clients qui souscrivent à un dépôt à terme ont tendance à avoir un nombre plus élevé de contacts précédents.


### **Conclusions Générales :**
- Le nombre de contacts précédents et le taux de variation de l'emploi semblent avoir une influence sur la souscription.
- Les variables économiques telles que le taux Euribor à 3 mois et le nombre d'employés semblent également jouer un rôle.
